- mode

    - by default the mode is : **r** read mode

- if we wants to create file and write any information then we need to change the mode

'r'       open for reading (default)

'w'       open for writing, truncating the file first

'x'       create a new file and open it for writing

'a'       open for writing, appending to the end of the file if it exists

'b'       binary mode

```python
# first 
file_name="pushpa.txt"
file=open(file_name,mode='w')
file.write('movie hit')
```

```python
file_name='pushpa2.txt'
with open(file_name,mode='a') as file:
    file.write('one time watch')
    
```

```python
#Q3) get the lines which are starting with from (startwith)
file_name=r"C:\Users\MY PC\Documents\Full Data Science & AI\DataFiles\mbox-short.txt"
with open(file_name,mode='r',encoding='utf-8-sig') as file:
    
    from_line=[line for line in file if line.startswith('From')]   
from_line
```

```python
#Q4) Remove the \n at the end of the line strip
list2=[line.strip('\n') for line in from_line]
list2
```

```python
#Q5) extract the email id
for i in range(len(list2)):
    print(list2[i].split()[1])

for i in list2:
    print(i.split()[1])

emails=[i.split()[1] for i in list2]
emails
```
